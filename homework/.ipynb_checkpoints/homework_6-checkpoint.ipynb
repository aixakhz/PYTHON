```python
#2. van der Waals Gas
```

```python
#(CH3)2CO - Acetone
a=16.02
b=0.1124

V,P=np.meshgrid(np.linspace(10,30),np.linspace(1,10))
n=1
R=0.083144

T=((P+a*(n**2/V**2))*(V-n*b))/(n*R)

plt.figure()
plt.title('Temperature')
plt.imshow(T,extent=[10,30,1,10],cmap='hot',origin='lower',vmin=100,vmax=3350)
plt.xlabel('Volume')
plt.ylabel('Pressure')
plt.colorbar()
plt.show()

#I tried to follow the format from the 'plotting_data_demo' and 'advanced_plotting_demo'
```

```python
#3. Monte Carlo 'pi'
```

```python
N_val=1e3,1e4,1e5,1e6

def est_pi(N):
    pts=np.random.rand(int(N),2)
    dist=np.linalg.norm(pts, axis=1)
    pts_inside_circle=pts[dist<=1]
    pi_estimate=4*len(pts_inside_circle)/N
    return pi_estimate, pts, pts_inside_circle

for N in N_val:
    pi_estimate, pts, pts_inside_circle=est_pi(N)
    print(f"N={N}: Estimated pi={pi_estimate}")
    
    if N==1e4:
        plt.figure()
        plt.title('Monte Carlo')
        plt.scatter(pts[:,0], pts[:,1],color='thistle')
        plt.scatter(pts_inside_circle[:,0],pts_inside_circle[:,1],color='lightcoral')
        plt.xlabel('X')
        plt.ylabel('Y')
        plt.show()
#I had a friend not in this class help me out with this problem
```
